# BUBBLE CHARTS
## Here size of the bubble can signfy something 
## like av salary vs year scatter graph, we want show working people of the population

```python
import matplotlib.pyplot as plt
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
# normal scatter graph
plt.scatter(year,salary,s=100)
plt.show()
```

```python
plt.scatter(year,salary,s=population,c="b")
plt.show()
```

```python
# alpha defines the tansparency to opaqe by default 1
plt.scatter(year,salary,s=population,c="b",alpha=0.5)
#plt.title("Bubble chart")
plt.show()
```

```python
# making different color bubble
# edgecolor to color to edge of bubble
c=np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor="black")
#plt.title("Bubble chart")
plt.show()
```

```python
# adding to text each value say polpulation
# marker to change bubble type
c=np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor="black",marker="^")
#plt.title("Bubble chart")
for i in range(len(population)):
    plt.text(year[i]+0.1,salary[i]-0.1,population[i])
plt.show()
```

# PIE CHARTS

```python
sizes=[3,4,6,2]
plt.pie(sizes)
plt.show()
# this graph is no exact circle
```

```python
# making circle by plt.axis("equal")
sizes=[3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
# changing colors
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
plt.pie(sizes,colors=colors)
plt.show()
```

```python
# adding labels
labels=["A",'B','C','D']
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
plt.pie(sizes,labels=labels,colors=colors)
plt.show()
```

```python
# Fill the graph with percentage or function
# use autopct
# if percentage "%.1f":shows upto 1 decimal;"%.2f":shows upto 2 decimal
# "%.1f%%":shows upto 1 decimal with percentage symbol,basically 1st % after f indicates next char is special 
labels=["A",'B','C','D']
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
#plt.title("pie")
plt.pie(sizes,labels=labels,colors=colors,autopct="%.2f%%")
plt.show()
```

```python
# let perform some fun f
```

```python
# Highlighting using explode 
explode=[0.1,0,0,0.2]
labels=["A",'B','C','D']
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%.2f%%")
plt.show()
```

```python
# Making counterclockwise using counterclock=False
explode=[0.1,0,0,0.2]
labels=["A",'B','C','D']
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
plt.pie(sizes,explode=explode,labels=labels,counterclock=False,colors=colors,autopct="%.2f%%")
plt.show()
```

```python
# Starting from some angle using startangle=100
explode=[0.1,0,0,0.2]
labels=["A",'B','C','D']
sizes=[3,4,6,2]
colors=["green","blue","orange","purple"]
plt.pie(sizes,explode=explode,labels=labels,startangle=100,counterclock=False,colors=colors,autopct="%.2f%%")
plt.show()
```

```python
plt.hist(b)
plt.show()
# no are clubed together
```

```python
# here width is (22-1)/10=2.1
# so first tower contains [1,3.1)
# 2nd tower contains [3.1,5.2)
```

```python
# We can set no. of bins
plt.hist(b,edgecolor="black",bins=20)
a=np.arange(23)
plt.xticks(a)
plt.show()
# width=21/20
```

```python
import numpy as np
import matplotlib.pyplot as plt
  
  
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES']
  
data = [23, 17, 35, 29, 12, 41]
  
  
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
  
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
  
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
  
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
  
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data, 
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode, 
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
  
# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
  
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
  
# show plot
plt.show()
```

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np
  
  
# Creating dataset
size = 6
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES']
  
data = np.array([[23, 16], [17, 23],
                 [35, 11], [29, 33],
                 [12, 27], [41, 42]])
  
# normalizing data to 2 pi
norm = data / np.sum(data)*2 * np.pi
  
# obtaining ordinates of bar edges
left = np.cumsum(np.append(0, 
                           norm.flatten()[:-1])).reshape(data.shape)
  
# Creating color scale
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(6)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9,
                              10, 12, 13, 15,
                              17, 18, 20 ]))
  
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7),
                       subplot_kw = dict(polar = True))
  
ax.bar(x = left[:, 0],
       width = norm.sum(axis = 1),
       bottom = 1-size,
       height = size,
       color = outer_colors, 
       edgecolor ='w',
       linewidth = 1,
       align ="edge")
  
ax.bar(x = left.flatten(),
       width = norm.flatten(),
       bottom = 1-2 * size,
       height = size,
       color = inner_colors,
       edgecolor ='w',
       linewidth = 1,
       align ="edge")
  
ax.set(title ="Nested pie chart")
ax.set_axis_off()
  
# show plot
plt.show()
```
